```python
# Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 data = pd.read_csv('1920matches.csv')
```

```python
data.head()
```

```python
data = data.drop('Div',axis=1)
```

```python
data.head()
```

```python
Teams = data.HomeTeam.unique()
```

```python
TeamList = {Team:[0] for Team in Teams}
```

```python
for row in data.itertuples():
    Home = row.HomeTeam
    Away = row.AwayTeam
    
    if row.FTHG > row.FTAG:
        TeamList[Home].append(3)
        TeamList[Away].append(0)
    
    elif row.FTHG < row.FTAG:
        TeamList[Home].append(0)
        TeamList[Away].append(3)
        
    else:
        TeamList[Home].append(1)
        TeamList[Away].append(1)
```

```python
TeamList['Norwich']
```

```python
Teams = data.HomeTeam.unique()
TeamLists = {Team : [0] for Team in Teams}

TeamGoals = {Goal : [0] for Goal in Teams}

```

```python
for row in data.itertuples():
    Home = row.HomeTeam
    Away = row.AwayTeam
    
    if row.FTHG > row.FTAG:
            TeamLists[Home].append(TeamLists[Home][-1]+3)
            TeamLists[Away].append(TeamLists[Away][-1]+0)
    
    elif row.FTHG < row.FTAG:
            TeamLists[Home].append(TeamLists[Home][-1]+0)
            TeamLists[Away].append(TeamLists[Away][-1]+3)
    
    else:
            TeamLists[Home].append(TeamLists[Home][-1]+1)
            TeamLists[Away].append(TeamLists[Away][-1]+1)
```

```python
Matchday = list(range(0,39))
```

```python
plt.plot(Matchday,TeamLists['Man United'],color='red',linewidth=2)
plt.plot(Matchday,TeamLists['Chelsea'],color='skyblue',linewidth=2)
```
