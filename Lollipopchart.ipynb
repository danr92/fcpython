```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
table = pd.read_csv('1920leaguetable.csv')
```

```python
fig = plt.figure(figsize=(10,7))
ax = plt.subplot()

plt.rcParams['font.family'] = 'DejaVu Sans'

plt.hlines(y=np.arange(0,20),xmin=0,xmax=table['Pts'], color=table['Kit'])
plt.plot(table['Pts'], np.arange(0,20), "o")
plt.yticks(np.arange(0,20),table['Squad'])

plt.ylabel('Team')
plt.xlabel('Points')
plt.text(55, 14,'Premier League 19/20',size=18,fontweight='bold')



for side in ['right','left','top','bottom']:
        ax.spines[side].set_visible(False)
        

```
