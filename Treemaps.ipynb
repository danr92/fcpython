```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import squarify
```

```python
data = pd.read_csv("livplayers1920.csv")
```

```python
dataGoals = data[data['Gls']>1]
```

```python
dataGoals
```

```python
norm =matplotlib.colors.Normalize(vmin=min(dataGoals.Gls), vmax=max(dataGoals.Gls))
colors = [matplotlib.cm.Blues(norm(value)) for value in dataGoals.Gls]
```

```python
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 8)

squarify.plot(label=dataGoals.Player, sizes=dataGoals.Gls, color=colors, alpha=.6)
plt.title('Liverpool 19/20 Goals', fontsize=22,fontweight='bold')
plt.axis('off')
```

```python
dataAssists=data[data['Ast']>1]
dataAssists.Ast
```

```python
norm1 = matplotlib.colors.Normalize(vmin=min(dataAssists.Ast),vmax=max(dataAssists.Ast))
colors1 = [matplotlib.cm.Reds(norm1(value)) for value in dataAssists.Ast]

squarify.plot(label=dataAssists.Player, sizes=dataAssists.Ast, color=colors1,alpha=0.6)

fig1 = plt.gcf()
ax1 = fig1.add_subplot()
fig1.set_size_inches(15,8)

plt.axis('off')
```
